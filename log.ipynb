```python
def show_log(name):
    log = output.fetch_log(name)
    _, ax = plt.subplots(figsize=(12, 8))
    def plot(key):
        data = [(l["iteration"], l[key])  for l in log if key in l]
        data = map(list, zip(*data))
        ax.plot(*data, label=key)
    
    plot("main/loss")
    plot("main/accuracy")
    plot("validation/main/loss")
    plot("validation/main/accuracy")
    
    ax.legend()
    ax.set_xlabel("iteration")
```

```python
show_log("ResNet50V1_1805021715")
```
